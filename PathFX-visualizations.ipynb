```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# For resizing purposes
fig, ax = plt.subplots(figsize=(12, 6))

# Plot cardiovascular drug data
cardio_drugs_num_assoc = [109, 45, 48, 51, 53] # Number of associations
cardio_drugs_bh_score = [0.000221262, 0.000240709, 0.000250648, 0.00025497, 0.000264477] # BH score
plt.scatter(cardio_drugs_num_assoc, cardio_drugs_bh_score, c='red', label="Cardiovascular Drugs") # Plot and change the color 

# Plot endocrine drug data
endo_drugs_num_assoc = [96] # Number of associations
endo_drugs_bh_score = [0.000113169] # BH score
plt.scatter(endo_drugs_num_assoc, endo_drugs_bh_score, c='blue', label="Endocrine Drugs") # Plot and change the color

# Plot respiratory drug data
resp_drugs_num_assoc = [49, 2386, 49, 2386] # BH score
resp_drugs_bh_score = [0.00005570164, 0.000074826679, 0.0000657169, 0.0001315266] # Number of associations
plt.scatter(resp_drugs_num_assoc, resp_drugs_bh_score, c='green', label="Respiratory Drugs") # Plot and change the color

# Plot GI drug data
gi_drugs_num_assoc = [122] # Number of associations
gi_drugs_bh_score = [0.00005697356] # BH score 
plt.scatter(gi_drugs_num_assoc, gi_drugs_bh_score, c='purple', label="Gastrointestinal Drugs") # Plot and change the color

# Name and label the plot
plt.xlabel("Number of Associations in the Interactome")
plt.ylabel("Benjamini-Hochberg Score")
plt.title("Number of Associations for Drugs vs Benjamini-Hochberg Score")
plt.legend()
plt.show()
```
